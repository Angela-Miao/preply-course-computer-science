```python
# difference between virus and bacteria?
# Virus (as we know now) is not classified as "living." Viruses can't reproduce by itself. Viruses are smaller than bacteria.
# Bacter is living. Bacteria can reproduce by itself. Bacteria are bigger than viruses.
```

```python
virus_names = [
    "Vaccinia virus",
    "Paramyxovirus (mumps)",
    "Herpesvirus",
    "Orf virus",
    "Rhabdovirus",
    "T-even coliphage",
    "Flexuous-tailed phage",
    "Influenza virus",
    "Polyomavirus",
    "Picornavirus",
    "Tubulovirus"
]
```

```python
virus_shapes = [
    "Complex brick-shaped",  # Vaccinia virus
    "Spherical with surface projections",  # Paramyxovirus (mumps)
    "Icosahedral",  # Herpesvirus
    "Oval",  # Orf virus
    "Bullet-shaped",  # Rhabdovirus
    "Bacteriophage with an icosahedral head and a tail",  # T-even coliphage
    "Rod-shaped with a flexuous tail",  # Flexuous-tailed phage
    "Spherical",  # Influenza virus
    "Polyhedral",  # Polyomavirus
    "Small icosahedral",  # Picornavirus
    "Rod-shaped"  # Tubulovirus
]
```

```python
names = {}
for i,j in zip(virus_names, virus_shapes):
    print(i, "is", j)
    names[i] = j
names
```

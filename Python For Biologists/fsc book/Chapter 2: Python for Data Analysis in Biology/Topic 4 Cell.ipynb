```python
# total_devision = total_time // divsion_time
## formula of cell total = inital * (2** total_devision)
```

```python
def simulate_cell_division(initial_cells, division_time, total_time):
    td = total_time // division_time
    total = initial_cells * (td**2)
    return (total)
# Example usage:
print(simulate_cell_division(1, 1, 3))  # Should print 8
```

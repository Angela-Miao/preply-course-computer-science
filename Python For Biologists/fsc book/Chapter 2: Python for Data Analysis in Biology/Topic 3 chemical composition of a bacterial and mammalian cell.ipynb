create a Python function named is_fibrous that takes a protein sequence as input and returns True if the sequence contains more than 70% of either 'A' (Alanine), 'G' (Glycine), 'V' (Valine), or 'L' (Leucine) amino acids, which are often prevalent in fibrous proteins; otherwise, it returns False. Implement this function in python.

```python
def is_fibous(x):
    total_length = len(x)
    pl = 0
    for i in x:
        if i in ["A", "G", "V", "L"]:
            pl = pl + 1 

    p = pl/total_length
    if p > 0.7:
        return(True, p*100 )
    else:
        return(False,p*100) 
```

```python
SEQ = 'ACGTGHISKLODACTGAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAACGFHSEJJJDD'
is_fibous(SEQ)
```

Nucleic acids were first isolated in 1869 by F. Miescher from the nuclei of pus cells. Due to their isolation from nuclei and their acidic nature, they were named nucleic acids. Nucleic acids are of two types, deoxyribonucleic acid or DNA and ribonucleic acid or RNA. DNA occurs in chromosomes, in the nuclei of the cells and in much lesser amounts in mitochondria and chloroplasts. RNA is present in the nucleolus, in the ribosomes, in the cytosol and in smaller amounts in other parts of the cell.

In a faraway land, there lived an enthusiastic young geneticist named Alex. Alex worked in the grand Genetic Library, a place where the genetic codes of every creature were stored. These codes were written in a language of just four letters: A for Adenine, G for Guanine, T for Thymine, and C for Cytosine. The arrangement of these letters determined the traits of every living thing in the land.
One day, a mysterious plant was brought to the Genetic Library. It had flowers that changed colors throughout the day and roots that glowed in the dark. Fascinated, Alex decided to decode its DNA sequence to understand the magic behind these phenomena.


The plant's DNA was a long string of letters, but Alex noticed that certain patterns in the sequence seemed to correlate with the unique traits of the plant. Alex hypothesized that by identifying and counting these patterns, they could unlock the secrets of the plant's magical traits.


Your Challenge:


Alex has identified several short DNA patterns that seem to be associated with the color-changing flowers and glowing roots. Your task is to write a Python function that takes a long string of DNA and a smaller string representing a DNA pattern and counts how many times that pattern appears in the DNA sequence.

```python
import re
def count_dna_pattern(dna_sequence: str, pattern: str) -> int:
    # Your code here
    return len(re.findall(pattern, dna_sequence))

```

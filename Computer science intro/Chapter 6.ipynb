```python
import math
import matplotlib.pyplot as plt
x = range(0,100)
y = []
functions = [#math.acos,
math.acosh,
#math.asin,
#math.asinh,
#math.atan,
#math.atan2,
#math.atanh,
math.cos,
math.cosh,
#math.tan,
#math.tanh
 ]
for function in functions:
    for i in x:
        print(i)
        y.append(function(i/101) )


    plt.plot(x,y)
    plt.title(str(function))
    plt.show()
    y = []
```

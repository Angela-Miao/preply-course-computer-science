```python
import numpy as np

# Data
aboce_temp = np.array([10, 15, 20, 25, 35, 40, 50, 55, 60])
aboce_taste = np.array([0, 1, 4, 7, 10, 8, 5, 3, 1])

# Create the Vandermonde matrix
X = np.column_stack([aboce_temp**2, aboce_temp, np.ones_like(aboce_temp)])

# Use least squares to find coefficients
coeffs, _, _, _ = np.linalg.lstsq(X, aboce_taste, rcond=None)

# Extract coefficients
a, b, c = coeffs

# Print the equation
print(f"The fitted parabolic equation is: y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## plot sing wave
x = np.arange(1,20,0.01)
y = np.sin(x)
plt.plot(x,y)
```

```python
# plot cos
y = np.cos(x)
plt.plot(x,y)
```

```python
# plot tan
y = np.tan(x)
plt.plot(x,y)
```

```python
# plot logarithmic
y = np.log(x)
plt.plot(x,y)
```

```python
# plot exponential
y = np.exp(x)
plt.plot(x,y)
```

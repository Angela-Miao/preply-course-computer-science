```python
#Define the parameters for the heartbeat simulation duration = 10  # Duration of the simulation in seconds 
sampling_rate = 1000  # Number of samples per second (1 kHz) 
heartbeat_rate = 75  # Heartbeat rate in beats per minute (BPM) 
heartbeat_amplitude = 1.0  # Amplitude of the heartbeat waveform
```

```python
# Calculate the number of samples 
duration = 5
num_samples = int(duration * sampling_rate)
```

```python
import numpy as np
```

```python
time = np.linspace(0, duration, num_samples, endpoint=False)
time
```

```python
heartbeat_frequency = heartbeat_rate / 60.0  # Convert BPM to Hz
heartbeat_waveform = heartbeat_amplitude * np.sin(2 * np.pi * heartbeat_frequency * time)
```

```python
import matplotlib.pyplot as plt
plt.plot(time, heartbeat_waveform)
```

```python
noise_amplitude = 0.01
heartbeat_waveform_with_noise = heartbeat_waveform + noise_amplitude * np.random.randn(num_samples)
```

```python
plt.plot(time, heartbeat_waveform_with_noise)
```

```python
import scipy
```

```python
pks,*_ = scipy.signal.find_peaks(heartbeat_waveform_with_noise)
```

```python
pks
```

```python
plt.plot(time, heartbeat_waveform_with_noise)
plt.plot(time[pks], heartbeat_waveform_with_noise[pks],'x')
```

```python
pks,*_ = scipy.signal.find_peaks(heartbeat_waveform_with_noise,height=0.95,distance=777)
plt.plot(time, heartbeat_waveform_with_noise)
plt.plot(time[pks], heartbeat_waveform_with_noise[pks],'x')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hyperbola
x = np.linspace(-10, 10, 100)
y = np.sqrt((x - 2)**2 / 9 - 1)
y_neg = -np.sqrt((x - 2)**2 / 9 - 1)

# Plotting
plt.plot(x, y, label="Hyperbola")
plt.plot(x, y_neg)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Hyperbola")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Circle
radius = 2
center = (1, -2)

# Generate theta values from 0 to 2*pi
theta = np.linspace(0, 2*np.pi, 100)

# Compute x and y coordinates of the circle
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

# Plotting
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Circle")
plt.axis("equal")  # Set equal scaling on both axes
plt.grid(True)
plt.show()

```

```python
import cv2
import numpy as np

# Load the image containing flowers
image = cv2.imread('plants.jpeg')

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper threshold values for pink color in HSV
lower_pink = np.array([150, 50, 50])
upper_pink = np.array([180, 255, 255])

# Create a binary mask by thresholding the image based on the pink color range
mask = cv2.inRange(hsv_image, lower_pink, upper_pink)

# Apply morphological operations to improve the mask
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Apply the mask to the original image to extract the pink flowers
pink_flowers = cv2.bitwise_and(image, image, mask=mask)

# Display the original image and the detected pink flowers
cv2.imshow('Original Image', image)
cv2.imshow('Detected Pink Flowers', pink_flowers)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the image containing flowers
image = cv2.imread('plants.jpeg')

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper threshold values for yellow color in HSV
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([40, 255, 255])

# Create a binary mask by thresholding the image based on the yellow color range
mask = cv2.inRange(hsv_image, lower_yellow, upper_yellow)

# Apply morphological operations to improve the mask
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Apply the mask to the original image to extract the yellow flowers
yellow_flowers = cv2.bitwise_and(image, image, mask=mask)

# Display the original image and the detected yellow flowers
cv2.imshow('Original Image', image)
cv2.imshow('Detected Yellow Flowers', yellow_flowers)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Initialize the video capture
video = cv2.VideoCapture(0)  # 0 corresponds to the default camera

# Loop over each frame from the camera
while True:
    # Read the current frame from the camera
    ret, frame = video.read()

    # Break the loop if the frame cannot be captured
    if not ret:
        break

    # Convert the frame to the HSV color space
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define the lower and upper threshold values for skin color in HSV
    lower_skin = np.array([0, 20, 70])
    upper_skin = np.array([20, 255, 255])

    # Create a binary mask by thresholding the frame based on the skin color range
    mask = cv2.inRange(hsv_frame, lower_skin, upper_skin)

    # Apply morphological operations to improve the mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    # Bitwise-AND the frame and the mask to extract the skin regions
    skin_regions = cv2.bitwise_and(frame, frame, mask=mask)

    # Display the original frame and the skin regions
    cv2.imshow('Original Frame', frame)
    cv2.imshow('Skin Regions', skin_regions)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
video.release()
cv2.destroyAllWindows()
```

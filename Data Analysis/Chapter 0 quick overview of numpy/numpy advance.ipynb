```python
import numpy as np
```

```python
fs = 500
t = np.arange(0,1,1/fs)
t
```

```python
# mmaking a clean signal
f = 5
clean_signal = np.sin(2*np.pi*f*t)
```

```python
import matplotlib.pyplot as plt
plt.plot(clean_signal,'.')
plt.title('clean signal')
```

```python
noise = np.random.normal(0,0.5,500)
plt.plot(noise)
```

```python
noise_signal = clean_signal + noise
```

```python
plt.plot(noise_signal)
```

```python
# we did not had data so we made our own data
# now we are going to remove noise from the singla
```

```python
#apply fft to the noise signal
fft = np.fft.fft(noise_signal)
plt.plot(fft)
```

```python
frequencies = np.fft.fftfreq(len(noise_signal),1/fs)
plt.plot(frequencies) 
```

```python
n = 5
fft[n:-n] = 0
```

```python
clean2 = np.fft.ifft(fft)
```

```python
plt.subplot(1,3,1)
plt.plot(clean_signal)

plt.subplot(1,3,2)
plt.plot(noise_signal)

plt.subplot(1,3,3)
plt.plot(clean2.real)
plt.show()

```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Sample list of strings
strings = bananas
# Count occurrences of each string
string_counts = Counter(strings)

# Extract string categories and their counts
categories = list(string_counts.keys())
counts = list(string_counts.values())

# Plotting the bar plot
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.bar(categories, counts, color='skyblue')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Bar Plot of String Occurrences')

# Rotating x-axis labels if they are long
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Sample list of strings
strings = bananas  # Replace 'bananas' with your actual list of strings

# Count occurrences of each string
string_counts = Counter(strings)

# Get the top 10 counted items
top_items = string_counts.most_common(10)  # Select top 10

# Extract string categories and their counts from the top 10 items
categories = [item[0] for item in top_items]
counts = [item[1] for item in top_items]

# Plotting the bar plot with a larger figure size
plt.figure(figsize=(10, 8))  # Increase figure size as needed
bars = plt.bar(categories, counts, color='skyblue')

# Adding labels and title
plt.xlabel('Categories', fontsize=12)  # Adjust label font size
plt.ylabel('Counts')
plt.title('Top 10 String Occurrences')

# Rotating x-axis labels if they are long and making them smaller
plt.xticks(rotation=45, fontsize=8)  # Adjust rotation and font size as needed

# Display the plot
plt.tight_layout()
plt.show()
```

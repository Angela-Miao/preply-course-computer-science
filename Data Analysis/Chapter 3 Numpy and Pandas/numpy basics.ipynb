```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given marks (excluding the outlier)
marks = [80, 85, 90, 90, 95, 95, 85, 80, 86, 88, 90, 87, 88, 84]

# Calculate mean and standard deviation
mean = np.mean(marks)
std_dev = np.std(marks)

# Generate data for the normal distribution curve
x = np.linspace(min(marks), max(marks), 100)
pdf = norm.pdf(x, mean, std_dev)

# Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Normal Distribution')

# Mark first, second, and third deviations
for i in range(1, 4):
    plt.fill_between(x, pdf, where=[(j >= mean - i * std_dev) and (j <= mean + i * std_dev) for j in x], alpha=0.3, label=f'{i} Deviation')

# Highlight the data points
plt.scatter(marks, norm.pdf(marks, mean, std_dev), color='red', label='Data Points')

# Add labels and legend
plt.title('Normal Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Probability Density Function')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a list with majority between 30 to 40 and minority between 70 to 80
minority_size = 20
majority_size = 1000

minor_values = np.random.uniform(70, 80, minority_size)
major_values = np.random.uniform(30, 40, majority_size)

data = np.concatenate([major_values, minor_values])

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Generate data for the normal distribution curve
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mean, std_dev)

# Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Normal Distribution')

# Mark the data points
plt.scatter(data, norm.pdf(data, mean, std_dev), color='red', label='Data Points')

# Add labels and legend
plt.title('Normal Distribution of Majority-Minority Data')
plt.xlabel('Values')
plt.ylabel('Probability Density Function')
plt.legend()

# Show the plot
plt.show()
```

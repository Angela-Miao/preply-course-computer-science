```python
import plotly.express as px

# Assuming you have a DataFrame named 'df'
# Example: df = pd.DataFrame({'col1': [...], 'col2': [...], 'col3': [...], 'RainToday': [...]})

# Scatter plot using Plotly
fig = px.scatter_3d(df, x='AvgPressure', y='avg_temp', z='avgHumidity', color='RainToday',
                    symbol='RainToday', opacity=0.7, size_max=10)

# Update marker symbols and legend
fig.update_traces(marker=dict(size=5),
                  selector=dict(mode='markers'))

# Set axis labels
fig.update_layout(scene=dict(xaxis_title='AvgPressure', yaxis_title='avg_temp', zaxis_title='avgHumidity'))

# Show the plot
fig.show()

```

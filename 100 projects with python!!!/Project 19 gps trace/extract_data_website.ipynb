```python
import json

file_path = 'gpstrace_node1.json'

# Initialize an empty list to hold all JSON objects
data_list = []

# Open and read the JSON file line by line
with open(file_path, 'r') as file:
    for line in file:
        try:
            data = json.loads(line)
            data_list.append(data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

# Print the list of JSON objects
print(data_list)
```

```python
for i in data_list:
    print(i)
```

```python
lat = []
log = []
for each in data_list:
    if each:
        lat.append(each.get('lat'))
        log.append(each.get('lon'))
    
```

```python
import matplotlib.pyplot as plt

plt.plot(lat, log)
plt.show()
```

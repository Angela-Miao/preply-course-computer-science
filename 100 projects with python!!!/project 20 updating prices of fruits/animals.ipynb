```python
data = open("animal.txt").readlines()
data
```

```python
animal = {}
for i in data:
    k,v =  i.split(' ') [0], (i.split(' ') [1])
    animal[k] = v
animal
```

```python
update_list = []
for i in data:
    update_list.append( i.replace("\n","").replace("kg","").replace("m/year","") ) 
    
```

```python
update_list
```

```python
d = {}
for i in update_list:
    k,v,v2 =  i.split(' ') [0], i.split(' ') [1],i.split(' ') [2]
    d[k+"_kg"] = v
    d[k+"_m/year"] = v2
d
```

```python
import pandas as pd
```

```python
data = pd.read_csv("1000_cryptos.csv")
data
```

```python

data.columns # whare are these??
```

```python
import matplotlib.pyplot as plt

# Sample data
data = data.iloc[-100:,:]
# Convert dates to a format that matplotlib can handle
dates = data.Date
open_price = data.Open
close_price = data.Close

plt.figure(figsize=(10, 5))
plt.plot(dates, open_price, label='Open Prices', marker='o')
plt.plot(dates, close_price, label='Close Prices', marker='x')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Open and Close Prices')
plt.legend()
plt.grid(True)
# rotate y label  90 degree
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
Mrandy = [8,9,10,11,12,13,14]
```

```python
Mrangela = [3,2,5,4,6,1,7]
```

```python
plt.plot(Mrandy,Mrangela,color = "#E4D7FA")
```

```python
plt.plot(Mrandy,Mrangela,color = "#E4D7FA",marker = "x")
```

```python
Msangela = [2,1,4,3,6,5,7]
```

```python
plt.plot(Mrandy,Mrangela)
plt.plot(Mrandy,Msangela)
plt.legend(["Mrangela","Msangela"])
```

```python
plt.subplot(2,1,1)
plt.plot(Mrandy,Mrangela)
plt.subplot(2,1,2)
plt.plot(Mrangela,Msangela)
```

```python
plt.subplot(2,2,1)
plt.plot(Mrandy,Mrangela)
plt.subplot(2,2,2)
plt.plot(Mrandy,Msangela)
plt.subplot(2,2,3)
plt.plot(Mrangela,Msangela)
plt.subplot(2,2,4)
plt.plot(Mrandy,Mrangela)
plt.savefig("MrandyangelaMs.png")
```
